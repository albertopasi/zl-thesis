Each `stream` object in the list `streamsX` is a dictionary, and typically contains the following important keys:

*   **`'info'`**: This is a dictionary holding metadata about the stream. It contains information such as:
    *   `'name'`: The name of the stream (e.g., 'actiCHampMarkers', 'actiCHamp-24020239', 'ZLT-logging').
    *   `'type'`: The type of data in the stream (e.g., 'Markers', 'EEG', 'misc', 'Logging').
    *   `'channel_count'`: The number of channels if it's a multi-channel stream.
    *   `'nominal_srate'`: The nominal sampling rate (samples per second).
    *   `'source_id'`: A unique identifier for the hardware source.
    *   `'created_at'`: Timestamp of when the stream was created.
    *   `'desc'`: A dictionary that can contain more detailed, custom descriptions or metadata for the stream, often including channel labels.

*   **`'time_series'`**: This is the actual data recorded in the stream. Its format depends on the stream type:
    *   For EEG or other numeric data, it's typically a NumPy array where rows are samples and columns are channels.
    *   For Marker streams, it's often a list of strings, where each string represents a marker event.

*   **`'time_stamps'`**: This is a NumPy array of timestamps, synchronized with the `'time_series'`. Each timestamp corresponds to a sample or a marker event.

```python
for i, stream in enumerate(streams1):
    print(f"\n=== Stream {i+1} ===")

    # Basic metadata
    name = stream['info']['name'][0]
    stype = stream['info']['type'][0]
    print("Name:", name)
    print("Type:", stype)

    # Time-series structure
    ts = stream['time_series']
    n_samples = len(ts)
    print("Samples:", n_samples)

    # Detect channels robustly
    if n_samples == 0:
        print("Channels: 0 (empty stream)")
    else:
        first_sample = ts[0]
        if isinstance(first_sample, (list, tuple)):
            print("Channels:", len(first_sample))
        else:
            print("Channels: 1")

        # Show first sample safely
        print("First sample:", first_sample)

    # Timestamps info
    print("Timestamps:", len(stream['time_stamps']))
```

```python
eeg_stream = streams1[4]
marker_stream = streams1[5]

eeg = np.array(eeg_stream['time_series'])
t = np.array(eeg_stream['time_stamps'])
marker_data = marker_stream['time_series']
marker_timestamps = marker_stream['time_stamps']
```

```python
# Define the custom serialization function
def numpy_to_python(obj):
    """Custom JSON serializer for NumPy data types"""
    if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
                        np.int16, np.int32, np.int64, np.uint8,
                        np.uint16, np.uint32, np.uint64)):
        return int(obj)
    elif isinstance(obj, (np.float_, np.float16, np.float32, np.float64)):
        return float(obj)
    elif isinstance(obj, (np.ndarray,)):
        return obj.tolist()
    raise TypeError(f"Object of type {obj.__class__.__name__} is not JSON serializable")

# Get the EEG info dictionary (Stream 5 is at index 4)
eeg_info = eeg_stream['info']

print("--- Full EEG Stream Info ---")
# Apply the custom serializer using the 'default' argument
# The user will need to run this command in their environment
print(json.dumps(eeg_info, indent=4, default=numpy_to_python))

sfreq = float(eeg_info['nominal_srate'][0])
duration_sec = eeg.shape[0] / sfreq

print(f"--- 1. Time Domain Analysis ---")
print(f"EEG Sampling Rate (Hz): {sfreq}")
print(f"Total Recording Duration: {duration_sec:.2f} seconds ({duration_sec / 60:.2f} minutes)")
print(f"Expected Samples (Duration * Fs): {duration_sec * sfreq:.0f}")
```

```python
# 2. Identify Channel Labels (Spatial Domain)
# The channel labels are nested deeply within the 'info' structure.
# Use a list comprehension to extract them.
channel_labels = [ch['label'][0] for ch in eeg_info['desc'][0]['channels'][0]['channel']]

print(f"--- 2. Channel Label Analysis ---")
print(f"Number of channel labels found: {len(channel_labels)}")
print("First 5 channel labels:", channel_labels[:5])
print("Last 5 channel labels:", channel_labels[-5:])  # What are AUX1 and AUX2?
```

```python
from collections import Counter

marker_stream = streams1[5]  # Stream 6 (ZLT-markers) is at index 5
marker_data = marker_stream['time_series']

# Convert the nested list structure (e.g., [['Start'], ['End']]) into a flat list of strings
# We use a try-except block to handle cases where marker_data might be a NumPy array.
try:
    # Flatten the data: marker_data is usually a list of single-item lists
    flat_markers = [item[0] for item in marker_data]
except:
    # Fallback assumption if it's a 1D NumPy array or already flat
    flat_markers = marker_data

# Count the frequency of each unique marker
marker_counts = Counter(flat_markers)

print(f"Total markers in ZLT-markers stream: {len(flat_markers)}")
print("\n--- Marker Frequency Analysis (Unique Markers and Counts) ---")

# Print the results sorted by frequency (most common first)
for marker, count in marker_counts.most_common():
    print(f"Marker: {marker:<30} | Count: {count}")
```
